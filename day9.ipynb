```python
student = {
    "name": "Vidula",
    "branch": "Data Science",
    "year": 1
}
#dictionary has key : branch
print(student["name"])
print(student.get("age"))   # returns None if key not found
student["collage"] = "Engineeering Collage" # cann add new item
print(student["collage"])
student["name"] = "vidu"    #can be updated
print(student["name"])
student.keys()
```

```python
student = {
    "name": "Vidula",
    "branch": "Data Science",
    "year": 1
}
#dictionary has key : branch
print(student["name"])
print(student.get("age"))   # returns None if key not found
student["collage"] = "Engineeering Collage" # cann add new item
print(student["collage"])
student["name"] = "vidu"    #can be updated
print(student["name"])
for key, value in student.items():  #looping
    print(key, ":", value)
```

```python
student = {
    1: {"name": "Vidula", "branch": "Data Science"},
    2: {"name": "Apurv", "branch": "CST"}
}
print(student[1]["name"])
print(student[2]["branch"]) #nested dictionary
```
